```python
#This took quite a long time, but the key to getting it was figuring out how to do operations properly, 
#and what type of object a .groupby() was. Feels good to have it done finally!

newdf = purchase_data.groupby(['Gender'])

meandf = newdf.mean()
sumdf = newdf.sum()
countdf = newdf.count()

avgdf = sumdf['Price']/countdf['Price']

perpersonavg = sumdf['Price']/gender_df['Gender']

#avgprice = newdf['Price'].sum()/newdf['Price'].count()

#pd.DataFrame({'Price':pd.to_numeric(newdf['Price'].sum()})


summary_table = pd.DataFrame({'Purchase Count':newdf['Price'].count(), 
                              'Total Purchase Value':newdf['Price'].sum(),
                              'Average Purchase Price': round(avgdf,2),
                             'Avg Total Purchase per Person': round(perpersonavg,2)})
summary_table



```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

group_names = ['<10', '10-15', '15-20', '20-25', '25-30', '30-35', '35-40', '40+']

remove_dupes["Age Range"] = pd.cut(remove_dupes["Age"], bins, labels=group_names)

purchase_data.head()
```

```python
group_df = remove_dupes.groupby("Age Range")['SN', 'Gender', 'Price' ].nunique().reset_index()

group_df.columns = ['Age Range', 'Total Count', 'Gender', 'Price']

group_df['Percentage of Players'] = round((group_df['Total Count']/ group_df['Total Count'].sum())*100,2)

#The binning wasn't too tough. The real difficulty here came from making sure that the .groupby() could be
#a proper datafram by using the .reset_index() function.

group_df2 = pd.DataFrame({'Age Range':group_df['Age Range'], 
                          'Total Count':group_df['Total Count'], 
                          'Percentage of Players':group_df['Percentage of Players']})

group_df2.set_index('Age Range', inplace=True)
group_df2
```

```python
purchase_data2 = purchase_data
purchase_data2["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

purchase_data3 = purchase_data2.groupby("Age Range")

purchase_data3['Price'].mean()

#note: the given example has a mistake in the age ranges, in that <10 is at the bottom. I changed this.

p_stats = pd.DataFrame({'Purchase Count': purchase_data3["Purchase ID"].count(),
                        'Average Purchase Price':round(purchase_data3['Price'].mean(),2),
                       'Total Purchase Value':purchase_data3['Price'].sum(),
                       'Average Total Purchase per Person': round(purchase_data3['Price'].sum()/group_df2['Total Count'],2),
                       
                       })

p_stats
```
