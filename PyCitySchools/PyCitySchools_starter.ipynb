```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_unique = pd.DataFrame({"Total Students":school_data_complete['size'].unique(),
              "school_name":school_data_complete['school_name'].unique(),
              "Total School Budget": school_data_complete['budget'].unique()
                   
                             })
```

```python
#There may be a quicker way to do this, but it seemed to work. 

#I see that "Dr. Richard Scott" is included as a student, and this may be implying some type of data cleaning.
#I also saw that the count of unique student names is shorter than the total student names. This implies one
#of two things: either students are improperly coded, potentially switching schools mid-year, or there 
#are students who share first and last name with other students. Given this ambiguity, I am performing all the
#below calculations on the full data. I could try to "clean" this, but it may not be accurate. I have no
#way of knowing which of these cases, or another, is true. 

avgmath = school_data_complete['math_score'].mean()
avgread = school_data_complete['reading_score'].mean()

school_data_complete['math_score'].count()
len(school_data_complete['student_name'])

mathcount = 0
for student in range(len(school_data_complete['student_name'])):
    if school_data_complete['math_score'][student] >= 70:
        mathcount +=1

mathcount    
```
